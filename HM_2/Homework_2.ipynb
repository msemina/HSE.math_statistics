```python
import numpy as np
import seaborn as sns
import numpy.random as rd
import matplotlib.pyplot as plt
import math
from IPython.display import display, Math
import scipy.stats as st
from numpy import arange
```

```python
#нормальное распределение
#h - масштабирование
def normal_pdf(x, mid=0, sigma=1, h=1):
    sqrt_two_pi = math.sqrt(math.pi * 2)
    return h*math.exp(-(x - mid) ** 2 / 2 / sigma ** 2) / (sqrt_two_pi * sigma)
```

```python
#Negative Binomial Distribution
cmap = plt.get_cmap("tab10").colors

plt.figure(figsize=(15, 15))

lam = 15


for k, r in enumerate([10, 50, 100, 500, 1000, 5000, 10000, 50000, 100000]):
    p = 1 - lam / (lam + r)
    c = [st.nbinom.pmf(i, r, p) for i in range(30)]
    plt.subplot(3, 3, k+1)
    plt.tight_layout()
    
    
    y = [normal_pdf(i, mid=lam, sigma=3.85, h=1) for i in arange (1,30,0.01)]
    
    plt.plot(list(arange (1,30,0.01)), y, color='black', label="Normal")
    plt.plot(list(range(30)), c, '.-', color=cmap[k], label="r = " + str(r))
    plt.title('NBinomial distr.')
    sb.set_style('whitegrid')
    plt.xlabel("k")
    plt.ylabel(r"$P(\xi _n=k)$")
    plt.legend()
plt.show()
```

```python
#Binomial Distribution
cmap = plt.get_cmap("tab10").colors

plt.figure(figsize=(15, 15))

lam = 15


for k, n in enumerate([20, 50, 100, 500, 1000, 5000, 10000, 50000, 100000]):
    
    p = lam / n 
    c = [st.binom.pmf(i, n, p, loc=0)  for i in range(30)]
    plt.subplot(3, 3, k+1)
    plt.tight_layout()
    
    
    y = [normal_pdf(i, mid=lam, sigma=3.85, h=1) for i in arange (1,30,0.01)]
    
    plt.plot(list(arange (1,30,0.01)), y, color='black', label="Normal")
    plt.plot(list(range(30)), c, '.-', color=cmap[k], label="n = " + str(n))
    plt.title('Binomial distr.')
    sb.set_style('whitegrid')
    plt.xlabel("k")
    plt.ylabel(r"$P(\xi _n=k)$")
    plt.legend()
plt.show()
```
